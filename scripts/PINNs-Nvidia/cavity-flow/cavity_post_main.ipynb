```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(50,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['Points_0'],df['Points_1'],df['u'],cmap='rainbow')

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('u_value')
```

```python
fig = plt.figure(figsize=(50,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['Points_0'],df['Points_1'],df['v'],cmap='rainbow')

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('v_value')
```

```python
fig = plt.figure(figsize=(50,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['Points_0'],df['Points_1'],df['v'],cmap='rainbow')
ax.view_init(elev=90, azim=-90)

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('v_value')
```

```python
fig = plt.figure(figsize=(50,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['Points_0'],df['Points_1'],df['p'],cmap='rainbow')
ax.view_init(elev=90, azim=-90)

ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')
ax.set_zlabel('p_value')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

scatter = ax.scatter(df['Points_0'],df['Points_1'],c=df['u'],cmap='rainbow', marker= 'o', s=50)
ax.set_xlabel('x_axis')
ax.set_ylabel('y_axis')

fig.colorbar(scatter, ax=ax, label= 'u_value')



```
