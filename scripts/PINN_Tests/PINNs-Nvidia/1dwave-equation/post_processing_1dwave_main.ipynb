```python
import pandas as pd
```

```python
df = pd.read_csv("results_int_points.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(50, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(df['Points:0'], df['t'], df['u'], cmap='rainbow')

ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')
ax.set_zlabel('u_axis')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

scatter = ax.scatter(df['Points:0'], df['t'], c=df['u'], cmap='rainbow', marker='o', s=100)

ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')

fig.colorbar(scatter, ax=ax, label='u_axis')
```
